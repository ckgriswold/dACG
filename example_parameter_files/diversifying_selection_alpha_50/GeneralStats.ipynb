```python
from ast import literal_eval
import matplotlib.pyplot as plt
```

```python
x=[]
with open('Data.out','r') as data_file: 
    for i in data_file: 
       x.append(literal_eval(i))
data_file.close()
```

```python
len(x)
```

```python
rep = 0

t = [i for i in range(len(x[rep]))]
plt.scatter(t,x[rep],color='k',s=1/1000)
plt.show()
```
